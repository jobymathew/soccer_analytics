```python
# How to divide into phases
# 1. Should have atleast 4 events
# 2. If there is a pause of 10 seconds between events
# 3. Lost ball to the other team, a phase can be divided
# 4. Less than 4 events would be considered if there is a shot
```

```python
def check_shot(phase):
    for data in phase:
        if data.subEventName == 'shot':
            return True
    return False

def check_time(previous, current):
    if (previous.matchPeriod == '1H' and current.matchPeriod == '2H'):
        return False
    elif (current.eventSec - previous.eventSec >= 10.0 ):
        return False
    return True

def get_phases(datas):
    phases = []
    phase = []
    for index, data in datas.iterrows():
#         print(row)
        if phase == []:
            phase.append(data)
        # Include if the same team
        elif phase[-1].teamId == data.teamId:
            # Seperating into phases if there is a time difference of 10 seconds between the events
            if check_time(phase[-1], data):
                phase.append(data)
            else:
                # If phase len > 3 or phase has a shot, include
                if len(phase) <= 3:
                    if check_shot(phase):
                        phases.append(phase)
                else:
                    print('Adding to the phase')
                    phases.append(phase)
                phase = []
        else:
            print('Team ID ', data.teamId)
            print('Row ', index)
            # If phase len > 3 or phase has a shot, include
            if len(phase) <= 3:
                if check_shot(phase):
                    phases.append(phase)
            else:
                print('Adding to the phase')
                phases.append(phase)
            phase = []
    return phases
            
        
```

```python
def convert_list_to_df(data):
    new_data = []
    for val in data:
        new_data.append(pd.DataFrame(val))
    return new_data
```

```python
# storing the same label values from clustering (AP)
def get_dict_label(labels, array):
    same_label = {}
    for i in range(len(labels)):
        if labels[i] in same_label:
            same_label[labels[i]].append(array[i])
        else:
            same_label[labels[i]] = []
            same_label[labels[i]].append(array[i])
    return same_label
```
